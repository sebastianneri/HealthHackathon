```python
import pandas as pd
import random
```

```python
df = pd.read_csv('municipios.csv', sep=';', skiprows=2, encoding='latin-1', dtype={'Clave de Entidad': str, 'Nombre de Entidad': str, 'Clave de Municipio':str, 'Nombre de Municipio':str} )
```

```python
df.drop(columns=['Nombre de Entidad','Nombre de Municipio'], inplace=True)
```

```python
df['Pob Riesgo'] = df['Clave de Municipio'].map(create_pob)
df
```

```python
df['Nuevos Casos'] = df['Clave de Municipio'].map(nuevos_casos)
df
```

```python
df['T Incidencia'] = (df['Nuevos Casos'] * 1000) / df['Pob Riesgo']
```

```python
df['T Incidencia'].max()
```

```python
def create_pob(x):
    if random.random() > 0.3:
        return random.randint(1500,2000)
    return random.randint(2000,4000)
```

```python
def nuevos_casos(x):
    if random.random() > 0.2:
        return random.randint(0,30)
    return random.randint(31,300)
```
